```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Temperature': [24, 27, 34, 32, 31, 29, 38, 36, 25, 28],
    'Ice_Cream_Sales': [300, 240, 450, 520, 280, 340, 430, 440, 290, 380],
    'Advertising_Budget': [50, 55, 70, 75, 60, 65, 80, 85, 90, 95],
    'Customer_Count': [130, 140, 135, 145, 170, 160, 150, 155, 165, 175]
}

df = pd.DataFrame(data)
sns.pairplot(df, diag_kind='kde', plot_kws={'alpha': 0.7})
plt.suptitle("Pair plot of ice cream sales data", y=1.02, fontsize=16)
plt.show()
```
