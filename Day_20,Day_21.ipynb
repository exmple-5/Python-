```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Create data
data = {
    "Study Hours": np.random.randint(1, 10, 50),
    "Marks": np.random.randint(30, 100, 50),
    "Screen Time": np.random.randint(1, 7, 50),
    "Travelling Time": np.random.randint(0, 3, 50),
    "Sleep Hours": np.random.randint(4, 10, 50),
}

# Create DataFrame
df = pd.DataFrame(data)

# Compute correlation matrix
corr = df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

# find cab prediction price(y) by considering x value(No.of km travelled 20) and rate per km is fixed one(m) as 2 and fixed price as that is fare is c value is 5
compare the prices of ola and uber which one low price and by considering below things
y=mx+c  fare price is 25 for ola, no. of km travelled(20),ola:base price is 50, uber:base price is 35, fare price: 50

```python

m = 2 
c = 5  
x = 20 
ola_base_fare = 50
ola_fare = m * x + c
total_ola_fare = ola_base_fare + ola_fare
uber_base_fare = 35
uber_fare = m * x + c
total_uber_fare = uber_base_fare + uber_fare

print(f"Ola Total Fare: {total_ola_fare}")
print(f"Uber Total Fare: {total_uber_fare}")

if total_ola_fare < total_uber_fare:
    print("Ola has a lower fare.")
else:
    print("Uber has a lower fare.")
```
